```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("df_performance.csv")
```

```python
df_select = df[['model', 'mode_nclus', 'avg_nclus', 'ri_best_clus', 'ari_best_clus']]
```

```python
display(df_select)
```

# Plot the average number of clusters

```python
g = sns.catplot(kind='box', x="model", y="avg_nclus", data=df).despine(
    left=False, right=False, top=False, bottom=False).set(
    xlabel='Model',
    ylabel='Average number of clusters'
)

g.refline(y=2, color='red')
plt.savefig('AVGncluster.pdf',dpi = 600)

```

# Plot the ARI of the best clustering according to Binder's loss

```python
g1 = sns.catplot(kind='box', x="model", y="ari_best_clus", data=df).despine(
    left=False, right=False, top=False, bottom=False).set(
        xlabel = 'Model',
        ylabel = 'ARI of Binder\'s estimates'
    )
plt.savefig('ARIS.pdf',dpi = 600)
```
